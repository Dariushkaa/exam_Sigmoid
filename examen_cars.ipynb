remove signs that do not carry important information

```python
# FOR ENGINE COLUMN: The function with which I leave only the fuel volume in the data, since the number of cylinders is in another attribute

import re

def extract_number(engine):
    # Разделение строки с помощью регулярного выражения
    match = re.search(r'[^,]+,\s*(\d+(\.\d+)?)', engine)

    if match:
        after_comma = match.group(1)   # Часть после запятой (число)

        # Преобразуем его в тип float
        number_float = float(after_comma)
    else:
         number_float = None

    return number_float

df['Engine'] = df.Engine.apply(extract_number)
df
```

```python
# For FuelConsumption Column: This attribute  csn become numeric, it should be cleaned of unnecessary rows
df["FuelConsumption100km"] = df.FuelConsumption.apply(lambda x : ' '.join(x.split(' ')[:1]))

df = df.drop(columns=['FuelConsumption'])

df['FuelConsumption100km'] = pd.to_numeric(df['FuelConsumption100km'], errors='coerce')

# For CylindersinEngine Column: This attribute  csn become numeric, it should be cleaned of unnecessary rows
df["CylindersinEngine"] = df.CylindersinEngine.apply(lambda x : ' '.join(x.split(' ')[:1]))
df['CylindersinEngine'] = pd.to_numeric(df['CylindersinEngine'], errors='coerce')


# # For Location Column: Separate the string by comma and create new attributes
df[['City', 'State']] = df['Location'].str.split(', ', expand=True)

# Удаление исходного признака Location
df = df.drop(columns=['Location'])

df
```

(P.S the more numeric data, the easier it is to process them)

```python
# For Doors and Seats columns : cleaned out

def extract_number(number):
    # Splitting a string using a regular expression
    match = re.search(r'\d', number)
    if match:
        return int(match.group(0))
    return None


df['Doors'] = df.Doors.apply(extract_number)

df['Seats'] = df.Seats.apply(extract_number)
df
```

```python

# For ColourExtInt column: Split the column into two new columns
df[['ExtColour', 'IntColour','None']] = df['ColourExtInt'].str.split(' / ', expand=True)

# Replacing '-' values with None
#df['ExtColour'] = df['ExtColour'].replace('-', np.nan)
#df['IntColour'] = df['IntColour'].replace('-', np.nan)


# Deleting the original ColourExtInt feature
df = df.drop(columns=['ColourExtInt'])
df = df.drop(columns=['None'])
df
```

```python

# Список цветов
colors = [
    'White', 'Silver', 'Grey', 'Black', 'Blue', 'Red', 'Green', 'Leather', 'Cloth', 'Charcoal',
    'Gold', 'Brown', 'Orange', 'Yellow', 'Beige', 'Burgundy',
    'Purple', 'Bronze', 'Turquoise', 'Other', 'Pink', 'Cream', 'Slate', 'Ebony','-','Other'
]

# Функция для проверки и замены значений
def check_and_replace(value, colors):
    if pd.isna(value):
        return value  # Возвращаем NaN без изменений

    # Разделяем строку на слова
    words = value.split()

    # Если строка состоит из одного слова и оно в списке цветов, возвращаем его
    if len(words) == 1 and words[0] in colors:
        return words[0]

    # Если в строке несколько слов, ищем совпадения с цветами
    for word in words:
        if word in colors:
            return word

    # Если совпадений нет, возвращаем 'Other'
    return 'Other'

df['IntColour'] = df['IntColour'].astype('str')
# Применяем функцию к столбцу 'ColourExtInt'
df['IntColour'] = df['IntColour'].apply(lambda x: check_and_replace(x, colors))
df['IntColour'].value_counts()


```

```python
df.drop(columns=['IntColour'], inplace=True)
```


This feature has 7755 missing values and 794 rows of 'Other' (set of rare colors) 50% of the data is undefined, hence we can remove them

```python
#For ExtColour Column
# Convert ExtColour column to string type
df['ExtColour'] = df['ExtColour'].astype('str')

# Define a function to replace the entire string with 'Other' if it contains numbers
def replace_numbers_with_other(text):
    if re.search(r'\d+', text):
        return 'Other'
    return text

# Applying the function to the ExtColour column
df['ExtColour'] = df['ExtColour'].apply(replace_numbers_with_other)

print(df)
```

```python
# Функция для замены NaN модой
def fillna_with_mode(df, columns):
    for column in columns:
        mode_value = df[column].mode()[0]  # Вычисление моды (наиболее часто встречающегося значения)
        df[column] = df[column].fillna(mode_value)  # Замена NaN на моду

# Список признаков для замены NaN на моду
features_to_fill = ['Engine', 'FuelConsumption100km', 'CylindersinEngine', 'Doors', 'Seats']

# Применение функции к DataFrame
fillna_with_mode(df, features_to_fill)
```

For the columns Engine, FuelConsumption, CylindersinEngine, Doors, and Seats, we fill the missing values with the mode obtained from the given data of each column

```python
# Function to perform feature extraction based on specific bins and categories

def feature_extraction(df):
    # Categorize 'Price' based on specified bins
    quantiles = df['Price'].quantile([0, 0.25, 0.5, 0.75, 1.0]).values
    df['PriceLevel'] = pd.cut(x=df['Price'],
                              bins= quantiles,
                              labels=['Low', 'Medium-Low', 'Medium-High', 'High'])

    # Categorize 'BloodPressure' based on specified bins
    df['KilometresLevel'] = pd.cut(x=df['Kilometres'],
                                     bins= [1, 100000, 200000, 300000, 400000, 550000],
                                     labels=['0-100k', '100k-200k', '200k-300k', '300k-400k', '400k-500k'])

    # Categorize 'Glucose' based on specified bins
    df['ConsumptionLevel'] = pd.cut(x=df["FuelConsumption100km"],
                               bins = [0, 5, 10, 15, 20, np.inf],
                               labels = ['Very Efficient', 'Efficient', 'Moderate', 'Less Efficient', 'Most Efficient'])



    # Perform one-hot encoding for the categorical features
    df = pd.get_dummies(df, columns=['PriceLevel', 'KilometresLevel', 'ConsumptionLevel'])

    return df

# Apply feature extraction to the DataFrame 'df'
df = feature_extraction(df)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def regression_models_predictions(X_train, y_train, X_test, y_test):
    # Initializing models
    models = {
        "Linear Regression": LinearRegression(),
        "Decision Tree": DecisionTreeRegressor(),
        "Random Forest": RandomForestRegressor(),
        "Support Vector Regressor": SVR()
    }

    predictions = {}
    results = []
    feature_importances = {}

    # Training models and making predictions
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        predictions[name] = y_pred
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        results.append((name, mse, mae, r2))

        # Сохранение важности признаков
        if hasattr(model, 'feature_importances_'):
            feature_importances[name] = model.feature_importances_
        elif hasattr(model, 'coef_'):
            feature_importances[name] = model.coef_

    # Создание таблицы предсказаний
    predictions_df = pd.DataFrame(predictions)
    predictions_df['Actual'] = y_test
    results_df = pd.DataFrame(results, columns=["Model", "Mean Squared Error", "Mean Absolute Error", "R^2 Score"])

    return predictions_df, results_df, feature_importances


predictions_df, results_df, feature_importances = regression_models_predictions(X_train, y_train, X_test, y_test)

print(results_df)

# Output of trait importance for Random Forest
if 'Random Forest' in feature_importances:
    feature_importance_df = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': feature_importances['Random Forest']
    }).sort_values(by='Importance', ascending=False)
    print(feature_importance_df)
```
